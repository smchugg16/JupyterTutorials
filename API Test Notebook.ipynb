```python
#creating api_runner 

def api_runner():
    global df 
#Original Sandbox Environment: 'https://sandbox-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest' 
    parameters = {
    'start':'1',
    'limit':'15',
    'convert':'USD'
}

headers = {
    'Accepts': 'application/json', 
    'X-CMC_PRO_API_KEY': '0ad53085-1cb2-4eb8-ad9e-3ffbd7e56509',
}



session = Session()
session.headers.update(headers)

try:
    response = session.get(url, params=parameters)
    data = json.loads(response.text)
    #print(data) 
except (ConnectionError, Timeout, TooManyRedirects) as e:
    print(e)
    

#keeping data in dataframe 
df2 = pd.json_normalize(data['data'])
df2['Timestamp'] = pd.to_datetime('now')
df_append = pd.DataFrame(df2)
df = pd.concat([df,df_append])

```

```python
#Running the API
import os 
from time import time 
from time import sleep 

for i in range(333):
    api_runner()
    print('API Runner Completed')
    sleep(60) #sleep for 1 minute 
exit()

#transfer into a csv file 
if not os.path.isfile(r'C:\Users\green\OneDrive\Documents\Python Scripts\API.csv'):
    df.to_csv(r'C:\Users\green\OneDrive\Documents\Python Scripts\API.csv', header='column_names')
else:
    df.to_csv(r'C:\Users\green\OneDrive\Documents\Python Scripts\API.csv', mode='a', header=False)
        
#Then to read in the file: 
df = pd.read_csv(r'C:\Users\alexf\OneDrive\Documents\Python Scripts\API.csv')
```
